All questions use: **Two-sample independent t-test**

```python
# Setup: Import required libraries
import numpy as np
from scipy import stats

def analyze_ttest(group1, group2, group1_name="One", group2_name="None"):
    """
    Perform two-sample independent t-test and generate APA-formatted output.
    
    Parameters:
    -----------
    group1 : list or array
        Data for group 1
    group2 : list or array
        Data for group 2
    group1_name : str
        Name for group 1 (default: "One")
    group2_name : str
        Name for group 2 (default: "None")
    """
    # Convert to numpy arrays
    g1 = np.array(group1)
    g2 = np.array(group2)
    
    # Calculate descriptive statistics
    mean1 = np.mean(g1)
    std1 = np.std(g1, ddof=1)  # Sample standard deviation
    n1 = len(g1)
    
    mean2 = np.mean(g2)
    std2 = np.std(g2, ddof=1)
    n2 = len(g2)
    
    # Perform independent samples t-test
    t_stat, p_value = stats.ttest_ind(g1, g2)
    
    # Calculate degrees of freedom
    df = n1 + n2 - 2
    
    # Determine significance
    is_significant = p_value < 0.05
    sig_text = "was" if is_significant else "was not"
    
    # Determine which group is higher
    if is_significant:
        if mean1 > mean2:
            comparison = f"Students with {group1_name} concentration scored significantly higher than students with {group2_name} concentration."
        else:
            comparison = f"Students with {group2_name} concentration scored significantly higher than students with {group1_name} concentration."
    else:
        comparison = ""
    
    # Print results
    print("="*70)
    print("DESCRIPTIVE STATISTICS")
    print("="*70)
    print(f"Group 1 ({group1_name} concentration):")
    print(f"  n = {n1}")
    print(f"  M = {mean1:.2f}")
    print(f"  SD = {std1:.2f}")
    print()
    print(f"Group 2 ({group2_name} concentration):")
    print(f"  n = {n2}")
    print(f"  M = {mean2:.2f}")
    print(f"  SD = {std2:.2f}")
    print()
    print("="*70)
    print("TEST RESULTS")
    print("="*70)
    print(f"Test: Two-sample independent t-test")
    print(f"t-statistic: {t_stat:.2f}")
    print(f"Degrees of freedom: {df}")
    print(f"p-value (two-tailed): {p_value:.3f}")
    print(f"Significance (α = 0.05): {'YES - SIGNIFICANT' if is_significant else 'NO - NOT SIGNIFICANT'}")
    print()
    print("="*70)
    print("="*70)
    print()
    print(f"A two-sample independent t-test was conducted to determine whether there ")
    print(f"was a significant difference in program outcomes between students with ")
    print(f"{group1_name} concentration and students with {group2_name} concentration. ")
    print(f"Students with {group1_name} concentration (n = {n1}) had a mean program ")
    print(f"outcome of {mean1:.2f} (SD = {std1:.2f}), while students with {group2_name} ")
    print(f"concentration (n = {n2}) had a mean program outcome of {mean2:.2f} ")
    print(f"(SD = {std2:.2f}). The difference {sig_text} statistically significant, ")
    print(f"t({df}) = {t_stat:.2f}, p = {p_value:.3f}. {comparison}")
    print()
    print("="*70)
    print("CONCISE VERSION")
    print("="*70)
    print()
    print(f"An independent samples t-test revealed {'a significant' if is_significant else 'no significant'} ")
    print(f"difference in program outcomes between students with {group1_name} concentration ")
    print(f"(M = {mean1:.2f}, SD = {std1:.2f}) and students with {group2_name} concentration ")
    print(f"(M = {mean2:.2f}, SD = {std2:.2f}), t({df}) = {t_stat:.2f}, p = {p_value:.3f}.")
    if comparison:
        print(comparison)
    print()
    print("="*70)

print("✓ Libraries loaded successfully!")
print("✓ analyze_ttest() function ready to use")
print("\nRun the cells below for each quiz question.")
```
