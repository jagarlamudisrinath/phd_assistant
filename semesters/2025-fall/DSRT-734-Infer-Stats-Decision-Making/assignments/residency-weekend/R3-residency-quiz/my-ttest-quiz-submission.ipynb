```python
# Setup: Import required libraries
import numpy as np
from scipy import stats
import pandas as pd

def analyze_ttest(group1, group2, group1_name="One", group2_name="None", question_num=1):
    """
    Perform two-sample independent t-test and generate detailed output.
    
    Parameters:
    -----------
    group1 : list or array
        Data for group 1
    group2 : list or array
        Data for group 2
    group1_name : str
        Name for group 1 (default: "One")
    group2_name : str
        Name for group 2 (default: "None")
    question_num : int
        Question number for labeling
    """
    # Convert to numpy arrays
    g1 = np.array(group1)
    g2 = np.array(group2)
    
    # Calculate descriptive statistics
    mean1 = np.mean(g1)
    std1 = np.std(g1, ddof=1)  # Sample standard deviation
    n1 = len(g1)
    
    mean2 = np.mean(g2)
    std2 = np.std(g2, ddof=1)
    n2 = len(g2)
    
    # Perform independent samples t-test
    t_stat, p_value = stats.ttest_ind(g1, g2)
    
    # Calculate degrees of freedom
    df = n1 + n2 - 2
    
    # Determine significance
    is_significant = p_value < 0.05
    sig_text = "was" if is_significant else "was not"
    
    # Determine which group is higher
    if is_significant:
        if mean1 > mean2:
            comparison = f"Students with {group1_name} concentration scored significantly higher than students with {group2_name} concentration."
        else:
            comparison = f"Students with {group2_name} concentration scored significantly higher than students with {group1_name} concentration."
    else:
        comparison = ""
    
    # Print results
    print("="*70)
    print(f"QUESTION {question_num} - ANALYSIS RESULTS")
    print("="*70)
    print()
    print("DESCRIPTIVE STATISTICS")
    print("-" * 70)
    print(f"Group 1 ({group1_name} concentration):")
    print(f"  Sample size (n) = {n1}")
    print(f"  Mean (M) = {mean1:.2f}")
    print(f"  Standard Deviation (SD) = {std1:.2f}")
    print()
    print(f"Group 2 ({group2_name} concentration):")
    print(f"  Sample size (n) = {n2}")
    print(f"  Mean (M) = {mean2:.2f}")
    print(f"  Standard Deviation (SD) = {std2:.2f}")
    print()
    print("TEST RESULTS")
    print("-" * 70)
    print(f"Test Used: Two-sample independent t-test")
    print(f"t-statistic: {t_stat:.3f}")
    print(f"Degrees of freedom (df): {df}")
    print(f"p-value (two-tailed): {p_value:.4f}")
    print(f"Significance level (α): 0.05")
    print(f"Result: {'SIGNIFICANT (Reject H₀)' if is_significant else 'NOT SIGNIFICANT (Fail to reject H₀)'}")
    print()
    print("APA-FORMATTED WRITE-UP")
    print("-" * 70)
    print()
    print(f"A two-sample independent t-test was conducted to determine whether there ")
    print(f"was a significant difference in program outcomes between students with ")
    print(f"{group1_name} concentration and students with {group2_name} concentration. ")
    print(f"Students with {group1_name} concentration (n = {n1}) had a mean program ")
    print(f"outcome of {mean1:.2f} (SD = {std1:.2f}), while students with {group2_name} ")
    print(f"concentration (n = {n2}) had a mean program outcome of {mean2:.2f} ")
    print(f"(SD = {std2:.2f}). The difference {sig_text} statistically significant, ")
    print(f"t({df}) = {t_stat:.2f}, p = {p_value:.3f}. {comparison}")
    print()
    print("="*70)
    
    # Return values for manual verification
    return {
        'mean1': mean1, 'std1': std1, 'n1': n1,
        'mean2': mean2, 'std2': std2, 'n2': n2,
        't_stat': t_stat, 'p_value': p_value, 'df': df,
        'significant': is_significant
    }

print("✓ Libraries loaded successfully!")
print("✓ analyze_ttest() function ready to use")
print("\nThis notebook is ready for quiz submission and PDF export.")
```

# Question 1 (10 Points)

## Problem Statement

**Research Question:** Is there a difference in means of program outcomes between students who have one concentration and students with no concentration?

### Data Provided

| One Concentration | No Concentration |
|-------------------|------------------|
| 61 | 60 |
| 62 | 61 |
| 62 | 61 |
| 63 | 62 |
| 63 | 63 |
| 65 | 65 |
| 70 | 70 |
| 77 | 77 |
| 77 | - |
| 79 | - |

**Sample sizes:** n1 = 10 (One concentration), n2 = 8 (No concentration)

## Hypotheses

**Null Hypothesis (H0):** There is no difference in mean program outcomes between students with one concentration and students with no concentration.
- H0: μ1 = μ2

**Alternative Hypothesis (H1):** There is a difference in mean program outcomes between the two groups.
- H1: μ1 ≠ μ2

**Significance Level:** α = 0.05 (two-tailed test)

## Test Selection

**Appropriate Test:** Two-sample independent t-test

**Justification:**
- Comparing means of two independent groups
- Continuous outcome variable (program outcome scores)
- Different subjects in each group (not paired data)
- Population standard deviation unknown

```python
# Question 1 Data
group1_one = [61, 62, 62, 63, 63, 65, 70, 77, 77, 79]
group2_none = [60, 61, 61, 62, 63, 65, 70, 77]

print("Group 1 (One concentration):", group1_one)
print("Group 2 (No concentration):", group2_none)
```

```python
# Run analysis
results_q1 = analyze_ttest(group1_one, group2_none, group1_name="one", group2_name="no", question_num=1)
```

## Notes & Reasoning

**Test Selection Justification:**
I selected the two-sample independent t-test because we're comparing program outcomes between two distinct groups of students: those with one concentration versus those with no concentration. These are independent groups (different students), the outcome variable is continuous (test scores), and we don't know the population standard deviation, making the t-test the appropriate choice.

**Data Observations:**
Looking at the raw data, both groups have similar ranges. The "One concentration" group has scores from 61-79 (n=10), while the "No concentration" group has scores from 60-77 (n=8). The data appears relatively normally distributed with no obvious outliers.

**Descriptive Statistics:**
After running the analysis, I found:
- One concentration: M = 67.90, SD = 7.20, n = 10
- No concentration: M = 64.88, SD = 5.84, n = 8
- Mean difference: 3.02 points (One concentration group slightly higher)

**Statistical Results:**
The t-test yielded t(16) = 0.96, p = 0.351. The p-value of 0.351 is much greater than our significance level of α = 0.05.

**Decision:**
I fail to reject the null hypothesis. There is insufficient statistical evidence to conclude that having one concentration affects program outcomes differently than having no concentration.

**Practical Interpretation:**
Although the "One concentration" group had a slightly higher mean (67.90 vs 64.88), this 3-point difference is not statistically significant. The p-value of 0.351 indicates there's a 35.1% chance we'd see a difference this large (or larger) due to random variation alone. In practical terms, students with one concentration performed about the same as students with no concentration.

**Assumptions Check:**
- Independence: Different students in each group ✓
- Normality: Sample sizes adequate, data appears reasonably normal ✓
- Random sampling: Assumed based on problem context ✓
