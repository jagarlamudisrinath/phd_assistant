# Question 1 (25 Points)

## Problem Statement

The attached dataset contains the distance a sample of people could throw a baseball and they could throw a football. Download the dataset, open it in JASP and determine if this dataset shows a correlation between the distance a person can throw a baseball and a football. Make sure to write your conclusion and APA statement.

**Dataset:** `baseball5.sav`

---

## Load Data

```python
# Load the baseball dataset
df_baseball, meta_baseball = pyreadstat.read_sav('baseball5.sav')

# Display dataset structure
print("Dataset Shape:", df_baseball.shape)
print("\nColumn Names:")
print(df_baseball.columns.tolist())
print("\nFirst few rows:")
print(df_baseball.head(10))
print("\nDescriptive statistics:")
print(df_baseball.describe())
```

## Pearson Correlation Analysis

```python
# Identify baseball and football columns
# Adjust column names based on actual data
baseball_col = df_baseball.columns[0]  # Assuming first column is baseball
football_col = df_baseball.columns[1]  # Assuming second column is football

print(f"Analyzing correlation between:")
print(f"  - {baseball_col}")
print(f"  - {football_col}")
print("="*60)

# Remove any rows with missing data
df_clean = df_baseball[[baseball_col, football_col]].dropna()

# Extract data
baseball_data = df_clean[baseball_col]
football_data = df_clean[football_col]

# Calculate Pearson correlation
r, p_value = pearsonr(baseball_data, football_data)

# Sample size
n = len(baseball_data)

# Descriptive statistics
baseball_mean = np.mean(baseball_data)
baseball_sd = np.std(baseball_data, ddof=1)
football_mean = np.mean(football_data)
football_sd = np.std(football_data, ddof=1)

print("\nDESCRIPTIVE STATISTICS")
print("="*60)
print(f"Sample Size: n = {n}")
print(f"\n{baseball_col}:")
print(f"  Mean = {baseball_mean:.3f}")
print(f"  SD   = {baseball_sd:.3f}")
print(f"\n{football_col}:")
print(f"  Mean = {football_mean:.3f}")
print(f"  SD   = {football_sd:.3f}")

print("\n\nPEARSON CORRELATION")
print("="*60)
print(f"Pearson r = {r:.3f}")
print(f"p-value   = {p_value:.4f}")
print(f"n         = {n}")

# Significance test
alpha = 0.05
if p_value < alpha:
    print(f"\nResult: SIGNIFICANT (p < {alpha})")
    print("Decision: Reject the null hypothesis")
    print("There IS a significant correlation between baseball and football throwing distances.")
else:
    print(f"\nResult: NOT SIGNIFICANT (p ≥ {alpha})")
    print("Decision: Fail to reject the null hypothesis")
    print("There is NO significant correlation between baseball and football throwing distances.")

# Interpret correlation strength
abs_r = abs(r)
if abs_r < 0.3:
    strength = "weak"
elif abs_r < 0.7:
    strength = "moderate"
else:
    strength = "strong"

direction = "positive" if r > 0 else "negative"
print(f"\nCorrelation Interpretation: {strength} {direction} relationship")
```

## APA-Formatted Conclusion

```python
# Generate APA conclusion
print("="*60)
print("APA-FORMATTED CONCLUSION")
print("="*60)
print()

if p_value < 0.001:
    p_text = "p < .001"
elif p_value < 0.01:
    p_text = "p < .01"
elif p_value < 0.05:
    p_text = "p < .05"
else:
    p_text = f"p = {p_value:.3f}"

if p_value < alpha:
    conclusion = f"""A Pearson correlation was conducted to examine the relationship between baseball 
throwing distance (M = {baseball_mean:.2f}, SD = {baseball_sd:.2f}) and football throwing 
distance (M = {football_mean:.2f}, SD = {football_sd:.2f}) for {n} participants. The analysis 
revealed a statistically significant {strength} {direction} correlation between the two variables, 
r({n-2}) = {r:.3f}, {p_text}. This indicates that individuals who can throw a baseball farther 
{'also tend to throw a football farther' if r > 0 else 'tend to throw a football shorter distances'}, 
suggesting a meaningful relationship between these throwing abilities."""
else:
    conclusion = f"""A Pearson correlation was conducted to examine the relationship between baseball 
throwing distance (M = {baseball_mean:.2f}, SD = {baseball_sd:.2f}) and football throwing 
distance (M = {football_mean:.2f}, SD = {football_sd:.2f}) for {n} participants. The analysis 
revealed no statistically significant correlation between the two variables, r({n-2}) = {r:.3f}, 
{p_text}. This suggests that the ability to throw a baseball farther is not significantly 
related to the ability to throw a football farther in this sample."""

print(conclusion)
```

# Question 5 (30 Points)

## Problem Statement

A teacher wants to know if there is a difference between student's mid-term exam scores and final exam scores in a business course. The attached dataset has both scores for each student in her class of 50 students. Save the following dataset to your computer and open it in JASP. Perform the correct t-test to see if the scores for the student's first exam are different than their scores on the final. Copy and paste your JASP output into your answer AND write your conclusion in APA format.

**Dataset:** `examscores24.sav`

**Correct Test:** Paired samples t-test (same students measured twice)

---

## Load Data

```python
# Load the exam scores dataset
df_exams, meta_exams = pyreadstat.read_sav('examscores24.sav')

# Display dataset structure
print("Dataset Shape:", df_exams.shape)
print("\nColumn Names:")
print(df_exams.columns.tolist())
print("\nFirst few rows:")
print(df_exams.head(10))
print("\nDescriptive statistics:")
print(df_exams.describe())
```

## Paired Samples t-Test

```python
# Identify midterm and final columns
# Adjust based on actual column names
midterm_col = df_exams.columns[0]  # Assuming first column is midterm
final_col = df_exams.columns[1]     # Assuming second column is final

print(f"Comparing:")
print(f"  - {midterm_col} (Measure 1)")
print(f"  - {final_col} (Measure 2)")
print("="*60)

# Remove any rows with missing data
df_clean = df_exams[[midterm_col, final_col]].dropna()

# Extract data
midterm_scores = df_clean[midterm_col]
final_scores = df_clean[final_col]

# Sample size
n = len(midterm_scores)

# Descriptive statistics
midterm_mean = np.mean(midterm_scores)
midterm_sd = np.std(midterm_scores, ddof=1)
midterm_se = midterm_sd / np.sqrt(n)

final_mean = np.mean(final_scores)
final_sd = np.std(final_scores, ddof=1)
final_se = final_sd / np.sqrt(n)

# Calculate difference scores
diff_scores = midterm_scores - final_scores
mean_diff = np.mean(diff_scores)
sd_diff = np.std(diff_scores, ddof=1)
se_diff = sd_diff / np.sqrt(n)

print("\nDESCRIPTIVE STATISTICS")
print("="*60)
print(f"Sample Size: n = {n}")
print(f"\n{midterm_col}:")
print(f"  Mean = {midterm_mean:.3f}")
print(f"  SD   = {midterm_sd:.3f}")
print(f"  SE   = {midterm_se:.3f}")
print(f"\n{final_col}:")
print(f"  Mean = {final_mean:.3f}")
print(f"  SD   = {final_sd:.3f}")
print(f"  SE   = {final_se:.3f}")
print(f"\nDifference ({midterm_col} - {final_col}):")
print(f"  Mean = {mean_diff:.3f}")
print(f"  SD   = {sd_diff:.3f}")
print(f"  SE   = {se_diff:.3f}")

# Perform paired samples t-test
t_stat, p_value = ttest_rel(midterm_scores, final_scores)
df = n - 1

# Calculate 95% confidence interval for mean difference
ci_95 = stats.t.interval(0.95, df, loc=mean_diff, scale=se_diff)

print("\n\nPAIRED SAMPLES T-TEST")
print("="*60)
print(f"t-statistic = {t_stat:.3f}")
print(f"df          = {df}")
print(f"p-value     = {p_value:.4f}")
print(f"Mean diff   = {mean_diff:.3f}")
print(f"95% CI      = [{ci_95[0]:.3f}, {ci_95[1]:.3f}]")

# Significance test
alpha = 0.05
if p_value < alpha:
    print(f"\nResult: SIGNIFICANT (p < {alpha})")
    print("Decision: Reject the null hypothesis")
    print("There IS a significant difference between midterm and final exam scores.")
else:
    print(f"\nResult: NOT SIGNIFICANT (p ≥ {alpha})")
    print("Decision: Fail to reject the null hypothesis")
    print("There is NO significant difference between midterm and final exam scores.")

# Calculate Cohen's d (effect size)
cohens_d = mean_diff / sd_diff
print(f"\nEffect Size (Cohen's d) = {cohens_d:.3f}")
if abs(cohens_d) < 0.2:
    effect_size_interpretation = "negligible"
elif abs(cohens_d) < 0.5:
    effect_size_interpretation = "small"
elif abs(cohens_d) < 0.8:
    effect_size_interpretation = "medium"
else:
    effect_size_interpretation = "large"
print(f"Interpretation: {effect_size_interpretation} effect")
```

## APA-Formatted Conclusion

```python
# Generate APA conclusion
print("="*60)
print("APA-FORMATTED CONCLUSION")
print("="*60)
print()

if p_value < 0.001:
    p_text = "p < .001"
elif p_value < 0.01:
    p_text = "p < .01"
elif p_value < 0.05:
    p_text = "p < .05"
else:
    p_text = f"p = {p_value:.3f}"

if p_value < alpha:
    if mean_diff > 0:
        direction_text = "higher on the midterm exam than on the final exam"
        implication = "indicating that student performance declined from midterm to final"
    else:
        direction_text = "higher on the final exam than on the midterm exam"
        implication = "indicating that student performance improved from midterm to final"
    
    conclusion = f"""A paired samples t-test was conducted to compare student exam scores on the midterm 
and final exams in a business course (n = {n} students). Results revealed a statistically significant 
difference between the two time points, t({df}) = {t_stat:.3f}, {p_text}. Students scored significantly 
{direction_text}, with a mean difference of {abs(mean_diff):.2f} points (95% CI [{ci_95[0]:.2f}, {ci_95[1]:.2f}]). 
The effect size was {effect_size_interpretation} (Cohen's d = {cohens_d:.2f}), {implication}."""
else:
    conclusion = f"""A paired samples t-test was conducted to compare student exam scores on the midterm 
and final exams in a business course (n = {n} students). Results revealed no statistically significant 
difference between the two time points, t({df}) = {t_stat:.3f}, {p_text}. Midterm scores (M = {midterm_mean:.2f}, 
SD = {midterm_sd:.2f}) were not significantly different from final exam scores (M = {final_mean:.2f}, 
SD = {final_sd:.2f}), suggesting that student performance remained relatively stable across the two exams."""

print(conclusion)
```

## Problem Statement

A company make three different products (A, B, and C). They took a sample of 50 random products to see if the average weight of the products are equal.

## Load Data

```python
# Load the product weights dataset
df_products, meta_products = pyreadstat.read_sav('Product33.sav')

# Display dataset structure
print("Dataset Shape:", df_products.shape)
print("\nColumn Names:")
print(df_products.columns.tolist())
print("\nFirst few rows:")
print(df_products.head(15))
print("\nDescriptive statistics:")
print(df_products.describe())
print("\nData structure info:")
print(df_products.info())
```

## Data Preparation

```python
# The data might be in wide format (3 columns: ProductA, ProductB, ProductC)
# or long format (1 column for product type, 1 for weight)
# We'll handle both cases

print(f"Dataset has {df_products.shape[1]} columns: {df_products.columns.tolist()}")

if df_products.shape[1] >= 3 and 'Product' not in str(df_products.columns[0]) and 'product' not in str(df_products.columns[0]).lower():
    # Wide format - separate columns for each product
    print("Data is in WIDE format")
    product_a = df_products.iloc[:, 0].dropna()
    product_b = df_products.iloc[:, 1].dropna()
    product_c = df_products.iloc[:, 2].dropna()
    
    # Create long format for easier analysis
    df_long = pd.DataFrame({
        'Product': ['A']*len(product_a) + ['B']*len(product_b) + ['C']*len(product_c),
        'Weight': list(product_a) + list(product_b) + list(product_c)
    })
else:
    # Long format - has product and weight columns
    print("Data is in LONG format")
    df_long = df_products.copy()
    
    # Find which column is Product and which is Weight
    product_col = None
    weight_col = None
    
    for col in df_long.columns:
        col_lower = str(col).lower()
        if 'product' in col_lower:
            product_col = col
        elif 'weight' in col_lower:
            weight_col = col
    
    # If not found by name, assume by position or data type
    if product_col is None or weight_col is None:
        # Check data types - product should be object/string, weight should be numeric
        for col in df_long.columns:
            if df_long[col].dtype == 'object' or df_long[col].dtype.name == 'category':
                product_col = col
            elif df_long[col].dtype in ['float64', 'int64', 'float32', 'int32']:
                weight_col = col
    
    print(f"Identified columns: Product='{product_col}', Weight='{weight_col}'")
    
    # Rename for consistency
    if product_col and weight_col:
        df_long = df_long[[weight_col, product_col]].copy()
        df_long.columns = ['Weight', 'Product']
    else:
        # Fallback: assume first is weight, second is product if we have exactly 2 columns
        if len(df_long.columns) == 2:
            df_long.columns = ['Weight', 'Product']
    
    print(f"Unique product identifiers: {sorted(df_long['Product'].unique())}")
    
    # The product identifiers might be 1,2,3 or 'A','B','C' or other values
    # Let's check and map if needed
    unique_products = sorted(df_long['Product'].unique())
    
    if len(unique_products) == 3:
        # Map to A, B, C if they're numeric or not already A,B,C
        if unique_products != ['A', 'B', 'C']:
            product_mapping = dict(zip(unique_products, ['A', 'B', 'C']))
            df_long['Product'] = df_long['Product'].map(product_mapping)
            print(f"Mapped products: {product_mapping}")
    
    # Extract separate arrays for each product
    product_a = df_long[df_long['Product'] == 'A']['Weight'].values
    product_b = df_long[df_long['Product'] == 'B']['Weight'].values
    product_c = df_long[df_long['Product'] == 'C']['Weight'].values

# Debug output
print("\n" + "="*60)
print("DEBUG: Data extracted for each product")
print("="*60)
print(f"Product A: {len(product_a)} values")
print(f"  First 5: {product_a[:5] if len(product_a) > 0 else 'EMPTY!'}")
print(f"  Mean: {product_a.mean():.2f}" if len(product_a) > 0 else "  Mean: N/A")
print(f"Product B: {len(product_b)} values")
print(f"  First 5: {product_b[:5] if len(product_b) > 0 else 'EMPTY!'}")
print(f"  Mean: {product_b.mean():.2f}" if len(product_b) > 0 else "  Mean: N/A")
print(f"Product C: {len(product_c)} values")
print(f"  First 5: {product_c[:5] if len(product_c) > 0 else 'EMPTY!'}")
print(f"  Mean: {product_c.mean():.2f}" if len(product_c) > 0 else "  Mean: N/A")
print()

print("\nSample sizes:")
print(f"Product A: n = {len(product_a)}")
print(f"Product B: n = {len(product_b)}")
print(f"Product C: n = {len(product_c)}")
print(f"Total: n = {len(product_a) + len(product_b) + len(product_c)}")
```

## Descriptive Statistics by Product

```python
# Calculate descriptive statistics for each product
products = {'A': product_a, 'B': product_b, 'C': product_c}

print("DESCRIPTIVE STATISTICS BY PRODUCT")
print("="*60)
print(f"{'Product':<10} {'N':<8} {'Mean':<12} {'SD':<12} {'SE':<12}")
print("-" * 60)

descriptives = {}
for name, data in products.items():
    # Safety check for empty or insufficient data
    if len(data) == 0:
        print(f"{name:<10} {'0':<8} {'N/A':<12} {'N/A':<12} {'N/A':<12}")
        print(f"  WARNING: Product {name} has no data!")
        descriptives[name] = {'n': 0, 'mean': np.nan, 'sd': np.nan, 'se': np.nan}
        continue
    elif len(data) == 1:
        # With only 1 data point, SD and SE are 0
        mean = data[0]
        descriptives[name] = {'n': 1, 'mean': mean, 'sd': 0.0, 'se': 0.0}
        print(f"{name:<10} {'1':<8} {mean:<12.3f} {'0.000':<12} {'0.000':<12}")
        print(f"  WARNING: Product {name} has only 1 data point!")
        continue
    
    # Normal calculation for n >= 2
    n = len(data)
    mean = np.mean(data)
    sd = np.std(data, ddof=1)
    se = sd / np.sqrt(n)
    
    descriptives[name] = {'n': n, 'mean': mean, 'sd': sd, 'se': se}
    print(f"{name:<10} {n:<8} {mean:<12.3f} {sd:<12.3f} {se:<12.3f}")

# Overall statistics (only if we have data)
all_weights = np.concatenate([product_a, product_b, product_c])
if len(all_weights) > 0:
    overall_mean = np.mean(all_weights)
    overall_sd = np.std(all_weights, ddof=1) if len(all_weights) > 1 else 0.0
    total_n = len(all_weights)
    
    print("-" * 60)
    print(f"{'Overall':<10} {total_n:<8} {overall_mean:<12.3f} {overall_sd:<12.3f}")
else:
    print("\nERROR: No data found in any product group!")
    total_n = 0
    overall_mean = np.nan
    overall_sd = np.nan

print()
```

## One-Way ANOVA

```python
# Perform one-way ANOVA
f_stat, p_value = f_oneway(product_a, product_b, product_c)

# Calculate degrees of freedom
k = 3  # number of groups
df_between = k - 1
df_within = total_n - k
df_total = total_n - 1

# Calculate sum of squares
# Between groups
ss_between = sum([len(data) * (np.mean(data) - overall_mean)**2 
                  for data in [product_a, product_b, product_c]])
# Within groups
ss_within = sum([np.sum((data - np.mean(data))**2) 
                 for data in [product_a, product_b, product_c]])
# Total
ss_total = ss_between + ss_within

# Mean squares
ms_between = ss_between / df_between
ms_within = ss_within / df_within

print("ONE-WAY ANOVA")
print("="*60)
print(f"{'Source':<15} {'SS':<12} {'df':<8} {'MS':<12} {'F':<10} {'p'}")
print("-" * 60)
print(f"{'Between Groups':<15} {ss_between:<12.3f} {df_between:<8} {ms_between:<12.3f} {f_stat:<10.3f} {p_value:.4f}")
print(f"{'Within Groups':<15} {ss_within:<12.3f} {df_within:<8} {ms_within:<12.3f}")
print(f"{'Total':<15} {ss_total:<12.3f} {df_total:<8}")
print()

# Significance test
alpha = 0.05
if p_value < alpha:
    print(f"Result: SIGNIFICANT (p < {alpha})")
    print("Decision: Reject the null hypothesis")
    print("At least one product has a significantly different mean weight.")
    print("\n→ POST HOC TESTS REQUIRED to determine which products differ.")
    run_posthoc = True
else:
    print(f"Result: NOT SIGNIFICANT (p ≥ {alpha})")
    print("Decision: Fail to reject the null hypothesis")
    print("No significant differences in mean weights among the three products.")
    print("\n→ Post hoc tests are NOT necessary (no significant differences found).")
    run_posthoc = False

# Calculate eta-squared (effect size)
eta_squared = ss_between / ss_total
print(f"\nEffect Size (η²) = {eta_squared:.3f}")
if eta_squared < 0.01:
    effect_interpretation = "negligible"
elif eta_squared < 0.06:
    effect_interpretation = "small"
elif eta_squared < 0.14:
    effect_interpretation = "medium"
else:
    effect_interpretation = "large"
print(f"Interpretation: {effect_interpretation} effect")
```

## Post-Hoc Tests (Tukey HSD)

```python
if run_posthoc:
    print("="*60)
    print("POST-HOC TESTS: Tukey HSD")
    print("="*60)
    print()
    
    if HSD_AVAILABLE:
        # Use statsmodels for Tukey HSD
        tukey = pairwise_tukeyhsd(endog=df_long['Weight'], 
                                   groups=df_long['Product'], 
                                   alpha=0.05)
        print(tukey)
        print()
        
        # Extract results for APA reporting
        tukey_df = pd.DataFrame(data=tukey.summary().data[1:], 
                                columns=tukey.summary().data[0])
        print("\nTukey Results Summary:")
        print(tukey_df.to_string(index=False))
    else:
        # Manual pairwise comparisons with Bonferroni correction
        print("Using pairwise t-tests with Bonferroni correction")
        print("(Note: Tukey HSD is preferred but not available)")
        print()
        
        comparisons = [
            ('A', 'B', product_a, product_b),
            ('A', 'C', product_a, product_c),
            ('B', 'C', product_b, product_c)
        ]
        
        bonferroni_alpha = alpha / len(comparisons)
        print(f"Bonferroni-adjusted alpha = {bonferroni_alpha:.4f}")
        print()
        print(f"{'Comparison':<15} {'Mean Diff':<12} {'t':<10} {'p':<10} {'Significant'}")
        print("-" * 60)
        
        for name1, name2, data1, data2 in comparisons:
            t, p = stats.ttest_ind(data1, data2)
            mean_diff = np.mean(data1) - np.mean(data2)
            sig = "Yes" if p < bonferroni_alpha else "No"
            print(f"{name1} vs {name2:<10} {mean_diff:<12.3f} {t:<10.3f} {p:<10.4f} {sig}")
else:
    print("\nPost-hoc tests not performed (ANOVA was not significant).")
```
